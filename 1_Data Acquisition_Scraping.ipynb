### Function to extract data from multiple links

```python
def get_movie_features(movie_page, filmmakers_page, details_page, companies_page):    
    
    soup2 = BeautifulSoup(movie_page, 'html.parser')
    soup3 = BeautifulSoup(filmmakers_page, 'html.parser')
    soup4 = BeautifulSoup(details_page, 'html.parser')
    soup5 = BeautifulSoup(companies_page, 'html.parser')
    
    #get movie title
    try:
        title_string= soup2.find('h1').findNext().text 
        title = title_string.split('(')[0].strip()
    except:
        title = 'unknown'
        
    #get gross (US $ Canada)
    try:
        gross_us_canada_string = soup2.find('div', class_= 'a-section a-spacing-small gross_usa_summary')\
                                    .find('div', class_='a-column a-span5 a-text-right a-span-last').text
        gross_us_canada = int(gross_us_canada_string.replace('$', '').replace(',', ''))
    except:
        gross_us_canada = 'unknown'

    #get gross (World)

    try:
        gross_world_string = soup2.find('div', class_= 'a-section a-spacing-small gross_world_summary').find('div', class_='a-column a-span5 a-text-right a-span-last').text
        gross_world = int(gross_world_string.replace('$', '').replace(',', ''))
    except:
        gross_world = 'unknown'


    #get movie rating
    try:
        rating= soup2.find('div', class_='a-box-inner a-padding-mini').text.replace(' ', '').replace('\n', '')
    except:
        rating = 'unknown'

    #get running time
    try:
        running_time_string = soup2.find('div', class_ = 'a-row a-spacing-top-micro truncate').find('span', id='running_time').text
        running_time = int(running_time_string.replace(' min', '').replace(' ', ''))
    except:
        running_time = 'unknown'

    #get genre 
    try:
        genre= soup2.find('div', class_ = 'a-row a-spacing-top-micro truncate').find('span', id = 'genres').text.replace(' ', '')
    except:
        genre= 'unknown'

    #get budget
    try:   
        budget_string = soup2.find('div', id= 'box_office_summary').find('div', class_ = 'a-column a-span5 a-text-right a-span-last').text
        budget = budget_string.replace('$', '').replace(',', '').strip()
    except:
        budget = 'unknown'

    #get reseale date
    try:
        release_date_string = soup2.find('div', class_= 'a-row a-spacing-base').find('a', class_="a-size- a-align- a-link-").text
    except:
        release_date_string = 'unknown'

    #get production company
    try:
        production_company= soup2.find('div', id= 'contacts').find('a', class_ = 'a-size- a-align- a-link-').text
    except:
        production_company= 'unknown'

    #get main actor
    try:
        main_actor= soup2.find('table', id= 'title_cast_sortable_table').find('span', class_='a-size-base-plus').text
    except:
        main_actor= 'unknown'

    #get main actor STARmeter
    try:
        main_actor_STARmeter_string= soup2.find('td', colspan  = '1', class_ = 'a-text-right').text.replace(' ', '').replace(',', '').replace('\n', '')
        main_actor_STARmeter = int(main_actor_STARmeter_string)
    except:
        main_actor_STARmeter = 'unknown'

    #get number of cast members
    try:
        number_of_cast_members_string = soup2.find('span', class_= 'a-size-mini a-color-secondary tab_subheading').text
        number_of_cast_members = int(number_of_cast_members_string.split(' ')[1])
    except:
        number_of_cast_members = 'unknown'

    #get distributor
    try:
        distributor= soup2.find('div', id= 'contacts').find('a', class_ = 'a-size- a-align- a-link-').findNext('a', class_="a-size- a-align- a-link-").text
    except:
        distributor= 'unknown'

    #get 'salesrep_trial', actually this would be distributor if it has a value. If this one is empty, distributor is accurate, 
    #if this one has a value, the value in distributor is actually of sales representative. 
    # add a special and descriptive try except

    try:
        salesrep_trial = soup2.find('div', id= 'contacts').find('a', class_ = 'a-size- a-align- a-link-').findNext('a', class_="a-size- a-align- a-link-")\
                        .findNext('a', class_ = 'a-size- a-align- a-link-').text
    except:
        salesrep_trial = 'unknown'

    ##### Features that require a change of page to 'filmmakers' page


    #get number of directors

    try:
        qty_directors = int(soup3.find('div', id = 'title_filmmakers_director_sortable_table_wrapper')\
                            .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_directors = 'unknown'

    #get number of writers
    try:
        qty_writers = int(soup3.find('div', id = 'title_filmmakers_writer_sortable_table_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_writers = 0

    #get number of producers
    try:
        qty_producers = int(soup3.find('div', id = 'title_filmmakers_producer_sortable_table_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_producers = 0

    #get number of art department crew members
    try:
        qty_art_department = int(soup3.find('div', id = 'title_filmmakers_art_department_sortable_table_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_art_department = 0

    #get number of sound department crew members
    try:
        qty_sound_department = int(soup3.find('div', id = 'title_filmmakers_sound_department_sortable_table_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_sound_department = 0

    #get number of special effects crew members
    try:
        qty_special_effects = int(soup3.find('div', id = 'title_filmmakers_special_effects_sortable_table_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_special_effects = 0

    #get number of visual effects crew members
    try:
        qty_visual_effects = int(soup3.find('div', id = 'title_filmmakers_visual_effects_sortable_table_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_visual_effects = 0

    #get number of camera and electrical department crew members
    try:
        qty_camera_and_electrical_department = int(soup3.find('div', id = 'title_filmmakers_camera_department_sortable_table_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_camera_and_electrical_department = 0

    #get number of casting department crew members
    try:
        qty_casting_department = int(soup3.find('div', id = 'title_filmmakers_casting_department_sortable_table_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_casting_department = 0

    #get number of editorial department crew members
    try:
        qty_editorial_department = int(soup3.find('div', id = 'title_filmmakers_editorial_department_sortable_table_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_editorial_department = 0

    #get number of music department crew members
    try:
        qty_music_department = int(soup3.find('div', id = 'title_filmmakers_music_department_sortable_table_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_music_department= 0

    #### Features that require a change of page to 'details' page

    #get number of filming locations
    try:
        qty_filming_locations= int(soup4.find('div', id = 'filming_locations').find('p', class_='a-spacing-small a-spacing-top-micro a-size-mini a-color-secondary').text\
                                .strip().split(' ')[0])
    except: 
        qty_filming_locations= 0

    #get number of soundtracks
    try:
        qty_soundtracks = int(soup4.find('div', id = 'soundtrack').find('p', class_='a-spacing-base a-size-mini a-color-secondary').text\
                          .strip().split(' ')[0])
    except:
        qty_soundtracks = 0

    #### Features that require a change of page to 'companies' page


    #get number of companies
    try:
        qty_companies= int(soup5.find('div', id = 'const_tabs').find('span', class_ = 'a-size-mini a-color-secondary tab_subheading').text.strip().split(' ')[0])                 
    except:
        qty_companies= 0

    #get number of production companies
    try:
        qty_production_companies = int(soup5.find('div', id = 'production_wrapper').find('span', class_ = 'additional_info')\
                                    .text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_production_companies = 0

    #get number of sales representatives
    try:
        qty_sales_representatives = int(soup5.find('div', id = 'sales_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_sales_representatives = 0

    #get number of distributors
    try:
        qty_distributors = int(soup5.find('div', id = 'distribution_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_distributors = 0

    #get number of special effects companies
    try:
        qty_special_effects_companies = int(soup5.find('div', id = 'specialEffects_wrapper')\
                        .find('span', class_ = 'additional_info').text.replace(' ', '').replace('(', '').replace(')', ''))
    except:
        qty_special_effects_companies = 0

    
 
    headers = ['title', 'gross_us_canada', 'gross_world', 'rating', 'running_time', 'genre', 'budget','release_date_string', 
               'production_company', 'main_actor',
               'main_actor_STARmeter', 'number_of_cast_members', 'distributor', 'salesrep_trial', 'qty_directors',
              'qty_writers', 'qty_producers', 'qty_art_department', 'qty_sound_department', 'qty_special_effects', 
                'qty_visual_effects', 'qty_camera_and_electrical_department', 'qty_casting_department',
              'qty_editorial_department', 'qty_music_department', 'qty_filming_locations', 'qty_soundtracks','qty_companies',
              'qty_production_companies', 'qty_sales_representatives', 'qty_distributors', 'qty_special_effects_companies']
    
    features = [title, gross_us_canada, gross_world, rating, running_time, genre, budget, release_date_string, 
               production_company, main_actor,
               main_actor_STARmeter, number_of_cast_members, distributor, salesrep_trial, qty_directors,
              qty_writers, qty_producers, qty_art_department, qty_sound_department, qty_special_effects, 
                qty_visual_effects, qty_camera_and_electrical_department, qty_casting_department,
              qty_editorial_department, qty_music_department, qty_filming_locations, qty_soundtracks, qty_companies,
              qty_production_companies, qty_sales_representatives, qty_distributors, qty_special_effects_companies]
    
    
        #Create movie dictionary and return
    movie_dict = dict(zip(headers, features))
   
    return movie_dict

get_movie_features(movie_page, filmmakers_page, details_page, companies_page)
```
